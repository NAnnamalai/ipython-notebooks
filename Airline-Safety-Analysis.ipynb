```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv", encoding="utf-8")
```

```python
df
```

```python
# Get top five fatalities caused by airlines in the period 2000-2014
most_fatalities = df.sort_values('fatalities_00_14', ascending=False).reset_index()
most_fatalities = most_fatalities[['airline', 'fatalities_00_14']]
most_fatalities['fatalities_00_14'] = most_fatalities['fatalities_00_14'].astype(int)
for ix, row in most_fatalities.head(5).iterrows():
    print row['airline'], row['fatalities_00_14']
```

```python
my_plot = most_fatalities.set_index('airline').head(15).plot(kind='bar')
```

```python
# Get top five fatal accidents caused by aurlines in the period 2000-2014
most_fatal_acc = df.sort_values('fatal_accidents_00_14', ascending=False).reset_index()
most_fatal_acc = most_fatal_acc[['airline', 'fatal_accidents_00_14']]
most_fatal_acc['fatal_accidents_00_14'] = most_fatal_acc['fatal_accidents_00_14'].astype(int)
for ix, row in most_fatal_acc.head(5).iterrows():
    print row['airline'], row['fatal_accidents_00_14']
```

```python
my_plot = most_fatal_acc.set_index('airline').head(15).plot(kind='bar')
```

```python
# Get most incidents caused by airlines
most_inc = df.sort_values('incidents_00_14', ascending=False).reset_index()
most_inc = most_inc[['airline', 'incidents_00_14']]
for ix, row in most_inc.head(5).iterrows():
    print row['airline'], row['incidents_00_14']
```

```python
my_plot = most_inc.set_index('airline').head(15).plot(kind='bar')
```

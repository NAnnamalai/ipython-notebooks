```python
import pandas as pd
%matplotlib inline
import matplotlib
```

```python
# https://raw.githubusercontent.com/datasets/browser-stats/master/data-extant.csv
df = pd.read_csv("https://raw.githubusercontent.com/datasets/browser-stats/master/data-extant.csv", encoding="utf-8", parse_dates=[0])
```

```python
df.head()
```

```python
df['year'] = df['Date'].dt.year
```

```python
df['year'].unique()
```

```python
browsers = set(df.columns) - set(['Date', 'year'])
browsers
```

```python
data = df.groupby(['year']).mean()
```

```python
data['Internet Explorer'].plot(kind='bar', x=data.index, title='Internet Explorer')
```

```python
data['Chrome'].plot(kind='bar', x=data.index, title='Chrome')
```

```python
data['Firefox'].plot(kind='bar', x=data.index, title='Firefox')
```

```python
data['Opera'].plot(kind='bar', x=data.index, title='Opera')
```

```python
data['Safari'].plot(kind='bar', x=data.index, title='Safari')
```

```python
data.plot(kind='bar', x=data.index, figsize=(12, 6), title='Browsers')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
labels = []
for col in data.columns:
    ax = data.plot(ax=ax, kind='line', x=data.index, y=col, figsize=(12, 6))
    labels.append(col)
lines, _ = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
plt.show()
```

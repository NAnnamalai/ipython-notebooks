```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv", encoding="utf-8")
```

```python
df.head(5)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('pandas_conditional.xlsx', engine='xlsxwriter')
```

```python
# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, index=False, sheet_name='Sheet1')
# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Sheet1']
```

```python
# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE', 'font_color': '#9C0006'})
```

```python
worksheet.conditional_format('A2:A101', {'type': '3_color_scale', 'format': format1})
worksheet.conditional_format('B2:B101', {'type': '2_color_scale', 'format': format1})
worksheet.conditional_format('C2:C101', {'type': 'data_bar', 'format': format1})
```

```python
# Create a new Chart object.
chart = workbook.add_chart({'type': 'column'})
chart.add_series({'values': '=Sheet1!$D$1:$D$101'})
chart.set_x_axis({
    'name': 'D',
    'name_font': {'size': 14, 'bold': True},
    'num_font':  {'italic': True },
})
# Insert the chart into the worksheet.
worksheet.insert_chart('F7', chart)
```

```python
writer.save()
```
